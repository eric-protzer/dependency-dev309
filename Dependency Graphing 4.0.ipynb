## edit the box below to choose your countries and years

```python
mycountry = ['VNM']
peers = ['CHN', 'LAO', 'THA', 'AGO', 'MMR', 'UZB', 'WSM']
years = [1996, 2001, 2006, 2011, 2016, 2021]
```

## don't edit below here

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
mycountries = peers + mycountry
data = pd.read_csv('data_out.csv')
data = data[data.year.isin(years)]
```

## graphs are down here

```python
#electricity

fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(data.loc[~data.ISO.isin(mycountries)]['year'], data.loc[~data.ISO.isin(mycountries)]['elec_mean'], c="gray")
ax.scatter(data.loc[data.ISO.isin(peers)]['year'], data.loc[data.ISO.isin(peers)]['elec_mean'], c="blue")
ax.scatter(data.loc[data.ISO == mycountry[0]]['year'], data.loc[data.ISO == mycountry[0]]['elec_mean'], c="red")
(_, caps, _) = ax.errorbar(data.loc[data.ISO == mycountry[0]]['year'], data.loc[data.ISO == mycountry[0]]['elec_mean'], yerr=data.loc[data.ISO == mycountry[0]]['elec_error'], fmt='o', capsize=5, c="red")
for cap in caps:
    cap.set_markeredgewidth(1)
ax.legend(['Rest of World', 'Peers', mycountry[0]], loc=9, bbox_to_anchor=(0.5, -0.15), ncol=2)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Dependency Coefficient', fontsize=12)
plt.title('Concentration of Exports by Electricity Dependence')
plt.show()
```
